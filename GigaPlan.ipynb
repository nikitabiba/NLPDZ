### Технологический стек
- Эмбэддинги: HuggingFaceEmbeddings(sberbank-ai/sbert_large_nlu_ru);
- LLM: GigaChat;
- Векторная БД: FAISS;
- Связь компонентов: LangChain;
- API: FastAPI;
- Интерфейс: streamlit.

### План
1) Чанковать данные с сохранением информации о структуре документов в метаданных;
2) Подбирать релевантные чанки с помощью BM25, FAISS Retriever, затем проводить переранжирование с помощью реранкера;
3) Написать системный промпт(содержимое прописано в тз);
4) Реализовать обращение к LLM(вместе с историей);
5) API с минимум одним эндпоинтом;
6) streamlit;
